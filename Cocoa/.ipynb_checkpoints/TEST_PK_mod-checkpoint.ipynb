```python
LOG_ALPHA_LIST = np.log(np.ones(6))
LOG_BETA_LIST  = np.log(np.ones(6))

LOG_BETA_LIST[0] = np.log(1.3)
LOG_BETA_LIST[3] = np.log(0.8)

test1 = LOG_ALPHA_LIST[k_indices][np.newaxis, :]
test2 = LOG_BETA_LIST[z_indices][:, np.newaxis]
test3 = test1+test2
print(test3.shape)

print(LOG_BETA_LIST)
print(LOG_ALPHA_LIST)
```

```python
k_add = np.zeros(len(k_indices))
z_add = np.zeros(len(z_indices))
```

```python
# z loop
for i in np.arange(6)+1:
    tmp = np.where(z_indices==i-1)[0]
    print(tmp)
    z_add[tmp] = LOG_BETA_LIST[i-1]
# k loop    
for i in np.arange(6)+1:
    tmp = np.where(k_indices==i)[0]
    k_add[tmp] = LOG_ALPHA_LIST[i-1]
```

```python
add_array = np.add.outer(z_add, k_add)
```
