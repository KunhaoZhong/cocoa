```python
lnpnl_before = np.load('baryon_test_before_sup_lnPNL.npy')
lnpnl_after  = np.load('baryon_test_after_sup_lnPNL.npy')

lnpnl_before = lnpnl_before.reshape(-1, 100).T
lnpnl_after = lnpnl_after.reshape(-1, 100).T
```

```python

for test_z in [0,20, 50]:
    plt.plot(lnpnl_before[test_z], label='pk nl before baryon', ls='solid')
    plt.plot(lnpnl_after[test_z],  label='pk nl after baryon', ls='-.')

plt.show()

plt.plot(lnpnl_after[0]/lnpnl_before[0])
plt.ylim(0,2)
```
